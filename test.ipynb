```python
# write code here for using geo.haversine() to get individual address distances
#THIS ONE WORKDS 
#I was trying to get the radius around each specific house but  kept getting runtime errors So I fetched a large amount of distances from vanouver area and then found the closest amenity to each  property (house).
reference_lat = 49.2827 
reference_lon = -123.1207  
radius = 10000  

#so we get the list of amenities
amenities = geo.get_specific_amenities(reference_lat, reference_lon, radius)

#go through over each property and calculate the distance to the nearest amenity
distances_to_amenities = []

for index, row in data_bc_single_family.iterrows():
    property_lat = row['latitude']
    property_lon = row['longitude']
    
    #LARGE NUM AS STARTING POINT
    min_distance = float('inf')
    
    #go through
    for amenity in amenities:
        amenity_lat = amenity['latitude']
        amenity_lon = amenity['longitude']
        
        #get the distance from place to place 
        distance = geo.haversine(property_lat, property_lon, amenity_lat, amenity_lon)
        
        #min distance 
        if distance < min_distance:
            min_distance = distance
    
    #keep min distance 
    distances_to_amenities.append(min_distance)

#add to data frame 
data_bc_single_family['distance_to_nearest_amenity'] = distances_to_amenities

#show 
print(data_bc_single_family[['streetAddress', 'addressLocality', 'price', 'distance_to_nearest_amenity']].head())  
```

```python
#TRYING TO INTERATE THROUGH THE AMENITIES WITHIN THE RADIUS OF THE GIVEN ADDRESSES 
radius = 1000

#list for amenities 
distances_to_amenities = []

#go through each property to get the distance from that property to the amenities 
for index, row in data_bc_single_family.iterrows():
    property_lat = row['latitude']  #get lat
    property_lon = row['longitude']  #get lon
    
    #get amenities 
    amenities = geo.get_specific_amenities(property_lat, property_lon, radius)
    
    #min distance
    min_distance = float('inf')
    
    # if no amenities within the radius then just say none
    if not amenities:
        distances_to_amenities.append(None) 
        continue

    #now go through the amenities to find the distance to each
    for amenity in amenities:
        amenity_lat = amenity['latitude']  #amenity latitude
        amenity_lon = amenity['longitude']  #amenity longitude
        
        #find the distance using haversine 
        distance = geo.haversine(property_lat, property_lon, amenity_lat, amenity_lon)
        
        #find the smallest distance of all  
        if distance < min_distance:
            min_distance = distance
    
    #add distance to the list 
    distances_to_amenities.append(min_distance)

#add the smallest distance to dataframe 
data_bc_single_family['distance_to_nearest_amenity'] = distances_to_amenities

#print 
print(data_bc_single_family[['streetAddress', 'addressLocality', 'price', 'distance_to_nearest_amenity']].head())
```

The code below is for normalizing data once we have it all cleaned up. We can ignore it for now and come back to it once we have clean data with all our columns

```python
features = ['price', 'property-beds', 'property-baths', 'Area']
scaler = MinMaxScaler()
data_scaled = sample.copy()
data_scaled[features] = scaler.fit_transform(sample[features])

data_scaled["price"] = 1 - data_scaled['price']
data_scaled['price'] = data_scaled['price'].map('{:.2f}'.format)
data_scaled['Garage'] = data_scaled['Garage'].map({'Yes': 1, 'No': 0})
data_scaled['Parking Spaces'] = data_scaled['Parking Spaces'].map({'Yes': 1, 'No': 0})
data_scaled.head()
```

```python
score_features = [
    'Price',
    'Bedrooms',
    'Bathrooms',
    'Square Footage',
    'Acreage',
    'Garage',
    'Parking'
]

weights = np.array([1/len(score_features)] * len(score_features))
data_scored = data_scaled.copy()
data_scored['Score'] = data_scaled[score_features].dot(weights)
data_scored.sort_values(by='Score', ascending=False, inplace=True)
data_scored
```
